```python
#input / outputs
x = data.drop('Life expectancy', axis='columns')
y = data['Life expectancy']
```

```python
#input
x
```

```python
#output
y
```

```python
#splitting Data
x_train , x_test, y_train , y_test = train_test_split(x, y, test_size = 0.2, random_state = 42 )
```

```python
display(x_train.shape, x_test.shape, y_train.shape , y_test.shape )
```

```python
#Fitting and training the model
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#testing model
y_pred = model.predict(x_test)
y_pred
print("TEST")
y_test
```

```python
#Checking Accuracy
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_squared_log_error, r2_score 
```

```python
print('r2 value :', r2_score(y_pred,y_test))
print('Mean Absolute error :',  mean_absolute_error(y_pred,y_test))
print('Mean Square error :',  mean_squared_error(y_pred,y_test))
print('Mean Square log error :',  mean_squared_log_error(y_pred,y_test))
```

model appears to perform very well based on these metrics. The high RÂ² value combined with low MAE, MSE, and MSLE indicates that it is not only fitting the training data well but is also likely to generalize effectively to unseen data. <br>
